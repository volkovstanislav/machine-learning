```python
def cleanhtml(raw_html):
    cleanr = re.compile("<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});")
    cleantext = re.sub(cleanr, "", raw_html)
    return cleantext

news['title'] = news['title'].apply(lambda x: cleanhtml(str(x)))
news['content'] = news['content'].apply(lambda x: cleanhtml(str(x)))
news.head(2)
```
